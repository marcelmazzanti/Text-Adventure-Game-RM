## A Brief Adventure Text Game

```python
"""
Rick & Morty is an animated TV show for adults started in 2013 which recently 
became very popular around the world. Every episode is based on a new adventure, 
each one with a humorous approach to pop-culture, current social issues and 
also nonsense physics. This game is meant to guide the user through some of 
these inventive Sci-Fi scenarios, reproducing scenes similar to those found in
the cartoon. Some variations have been made in regards of dialogues and plot,
although the game tries to be as coherent as possible to the traits of characters.

Trailer for the last season (4): https://www.youtube.com/watch?v=0kDgFKTJHAQ

Code by Marcello Mazzanti

"""
#imports for used functions
import time
import random
from sys import exit

#asking user name
#name variable
name = input (prompt= "What is your intergalactic name?\n\n").capitalize()

#making name a global variable to use for finals
global name

#display game context
print (f"""
Greetings {name}!

Today Rick is supposed to attend family therapy with Dr. Wong and needs to 
find an excuse to avoid it, he already turned himself into a pickle once, 
and he cannot take Morty because he has to go, luckily you are here {name}, 
or whoever you are...
""")

print (f"""
He is asking you to go on the next adventure with him! 
Rick says if you accept, this time you can be in charge of the episode 
and take all the decisions!
LET'S GO AROUND THE MULTIVERSE AND DEFEAT SOME OF RICK'S ENEMIES.

{'-' * 80}""")




#starting the game
input (prompt= f"""If you are ready to join him, press <Enter>\n\n""")


#question one STARTING checkpoint
#user path choise
def multiv():
    #creating a list for possible answers
    ans_a = ["A", "P", "Promortyus", "First", "1", "One", "Not ricklactica", "Not Ricklactica"]
    ans_b = ["B", "R", "Ricklactica", "Second", "2", "Two", "Not Promortyus", "Not promortyus"]
    
    #input first choice where game divides
    #define answer variable
    multi_ch = input (prompt= f"""
Rick takes the portal gun and asks you:
"Which part of the multiverse do you whish to visit?"

    A) Promortyus 
    B) Ricklactica\n
""").capitalize()
   
    #conditional statements directing to next scenes or repeting question
    #directs to MORTY scenario
    if multi_ch in ans_a:
        print (f"""
Rick: "Alright! I guess you are in the mood for some combat."\n{'-' * 80}""")
        time.sleep(3)
        prom_a()
        
    #dircets to RICK scenario
    elif multi_ch in ans_b:
        print (f"""
Rick: "Oh yes, good choice, you never know what to expect over there!"\n{'-' * 80}""")
        time.sleep(3)
        rick_b()
        
    #ask the user to repeat    
    else:
        print (f"""
What did you say? I still do not understand this Gazorpian dialect, 
try again with one of the two options...\n{'-' * 80}""")
        time.sleep(3)
        multiv()

        
#define retry function
def retry():
    print (f"""
You are about to die, but Rick is giving you the chance to move 
in the multiverse to a reality where all this has not happened yet.\n\n""")
    retry = input (prompt= "If you want to do that, press <Enter>)\n\n>" )
    print (f"""{'-' * 80}""")
    multiv()

#define final fail option
def fail():
    print (f"""\n{'-' * 80}
You are officially dead {name}, there is nothing to do about this, 
not even an anatomy park inside your body, that would smell very bad.



                        THE END.\n\n{'-' * 80}""")
    exit(0) 
    
    

#defining and starting FIRST option MORTY
def prom_a():
    
    #create a variable for loop trials SCARY TERRY
    #create lists for answers
    trial = 0
    
    scr_a = ["A", "Run Away", "Run Away", "Run", "Away", "Ruun", "first", "1", "One"]
    scr_b = ["B", "Hide", "Dont listen", "Second", "2", "Two" ]
    scr_c = ["C", "Talk", "Talk to him", "Convince him", "Convince", "Third", "3", "Three" ]
    
    #initial message SCARY TERRY
    print (f"""
Before all the fighting that awaits you, Rick has to take a nap,
because of course he is super drunk and depressed...
So he just uses a gas to put you to sleep as well not to bother him, 
but as soon as you fall asleep you notice you are in a very creepy dream with Rick...
{'-' * 80}""")
    time.sleep(4)
    #first prompt and input first scenario
    #creating a loop for hide option to be reapeted (3 times to escape) in SCARY TERRY
    while trial < 3:
        scary_t = input (prompt=f"""
You hear the voice of Scary Terry, the monster who kills his victims in their dreams,

He is screaming: "YOU CAN RUN, BUT YOU CAN'T HIDE"... What do you do??
    
    A) Run away!
    B) Hide, since when are we listening to the enemies insights?
    C) Talk to him, convince him he is a good guy, no need to always be scary.
    """).capitalize()
        
        #retry option SCARY TERRY
        if scary_t in scr_a:
            print (f"""
He moves fast inside dreams! He caught you easily...\n{'-' * 80}""")
            time.sleep(3)
            retry()
            break
            
        #hide and repeat option adding trials
        elif scary_t in scr_b:
            
            trial += 1
            print (f"""
IT WORKS! kind of, he is still looking for you...\n{'-' * 80}""")
            time.sleep(3)
            
             
        #win option to break loop and continue directly
        elif scary_t in scr_c:
            print (f"""
Wow, he actually listened... he also suggests where to continue your adventure, 
he knows a place where people have very weird dreams and he tells it to Rick\n{'-' * 80}""")
            break
            
        #restarting with input not recognized
        else:
            print (f"""
You keep talking with a weird language, Rick rewinds time to 
make you repeat and not sound stupid\n{'-' * 80}""")
                 
    prom_a2()
    
#defining function for next stage after loop
def prom_a2():
    
    #next stage message PURGE scene
    print (f"""Scary Terry is gone now...\n{'-' * 80}""")
    #waiting to continue
    time.sleep(3)
    
    print (f"""
The place where Rick is now taking you is a planet where apparently 
it is socially acceppted to PURGE only for one day every year, and guess what 
day is today.... PURGE DAY!\n{'-' * 80}
""" )
    time.sleep(4)
    #lists for PURGE answers
    prg_a = ["Try", "nonsense", "A", "First", "1", "Try to explain", "Try explain" "Try to explain the nonsense of this tradition", "Do not purge" ]
    prg_b = ["Suit", "Purge", "B", "Second", "Initiate unnecessarily violent suit", "Initiate suit", "Try suit"]
    
    #input and prompt for PURGE scene
    purg = input (prompt = f"""
You and Rick arrive when the purge has started already and find yourselves 
facing a group of aliens expressing all their rage, obviously they are
running toward you, Rick is asking how you want to manage this one...

    A) Try to explain the nonsense of this tradition.
    B) Initiate unnecessarily violent suit and purge them all.
    
""").capitalize()
    
    #conditional statements: one to continue and two to fail in PURGE scene
    if "Explain" in purg or "Dont purge" in purg or "Dont initiate" in purg or purg in prg_a:
        print (f"""
That would not work... They are so angry! They are now just tearing you apart.""")
        fail()
    #only viable option to survive
    elif "Initiate" in purg or "Suit" in purg or "Purge" in purg or purg in prg_b:
        print (f"""
Yes! Ok... Well, that got you weirdly excited...\n{'-' * 80}""")
    #case for different answers but ending the game this time, prompt user to play by rules.
    else:
        print (f"""
Rick has had enough of your confusing language, he is now bored...

He kills you and goes to family therapy...

Next time you better choose from the options he gives.\n{'-' * 80}""")
        fail()

    #final question for FIRST scenario
    print (f"""
You and Rick have made so much noise around the galaxy with all this purging,
that the Galactic Federation has now located Rick, they have been trying to 
capture rick for years... So now you are in real danger.\n{'-' * 80}""")
    time.sleep(5)
    #create list for final answer G. FED
    fin_lst = ["A", "Shoot", "Kill them", "Blast", "Destroy", "Use gun", "First", "1", "Use", "Use it"]
    
    #last input of First scenario
    final_a = input (prompt = f"""
The Army of the Galactic Federation is approaching...
Rick tells you that he has a very advanced guns to destroy the entire ship 
in just one blast, but these ships are like cities for them, there are entire 
families of aliens in there, they have kids too...

So what do you decide?

    A) Shoot
    B) Spare their lives
""").capitalize()
    
    #final conditional statement to win in FIRST scenario G. FED
    if final_a in fin_lst:
        print (f"""
{'-' * 80}
Reckless... But effective!
        
        
        
                         YOU WON {name}!!!
                         
                You have earned Rick's respect now.

{'-' * 80}""")
        exit(0)
        

    #answers different from list will make user fail    
    else:
        print (f"""{'-' * 80}\nYou have a good heart but...""")
        fail()
        

#define SECOND scenario function RICK
def rick_b():
    
    #initial message for SECOND option RICK
    print (f"""
You are travelling in the spaceship with Rick and you find an old box 
with a big button on it, Rick tells you it must be a Meeseeks box, 
if you press it a Meeseek will come out to make one wish come true, 
but there is a lot of rust on it and it might also be a Kirkland box, 
which is a very unproductive Meeseek. Because (actually) Rick wanted to go 
to Promortyus he tells you to guess which of the two will come out 
and press the button, but if you get it wrong he will have the wish... \n{'-' * 80}""")
  
    time.sleep(5)
    #creating a loop for the user to guess a random variable
    while True:
        #defining random variable and asking for guess
        box = random.choice(["Meeseek", "Kirkland"])
        guess = input (prompt= f"""
What is your guess?

    A) Meeseek
    B) Kirkland\n
""").capitalize()
        
        print (f"""It's a {box}!\n""")
        #option if guess is KIRKLAND
        if guess in box [0] or "A" in guess:
            #wrong guess
            if box == "Kirkland":
                print (f"""
Rick: "Oh well I get the wish, but this guy is useless anyways... let's keep going."\n{'-' * 80}""")
            #correct guess   
            else:
                print (f"""
Rick: "Ok you got it, let's keep going."\n{'-' * 80}""")
        #option if guess is MEESEEK        
        elif guess in box [1] or "B" in guess:
            #right guess
            if box == "Kirkland":
                print (f"""
Rick: You got it, well, whatever... Kirklands are super lame, 
no wish for you either, just a weirdo hanging around."\n{'-' * 80}""")
                
            #wrong guess and calling MORTY function 
            else:
                print (f"""
Rick: "Nice. I get the wish, I wish we were in Promortyus now!
Way more action over there!"

And PUFF!...
{'-' * 80}   """)
                prom_a()
        #repeting question         
        else:
            print (f"""
Rick: "That kind of guess is not valid, go for one of the two options."

...He rewinds time to see if you get it this time...
{'-' * 80}""")
            rick_b()

        break
        
    #initial message for TOXIC scene
    print (f"""
You now notice that Rick is being a bit too nice to you, and that is 
very suspicious... He says he recently has been in a very relaxing Spa where they 
removed the toxiicity from his body, as he keeps saying all nice things about it 
a Toxic version of Rick crashes into the spaceship and shouts:

"DID YOU THINK YOU COULD GET RID OF ME LIKE THIS?? I AM GOING TO KILL YOU AND 
MAKE THE WHOLE GALAXY TOXIC!!"

Rick: "Now I see! The Spa must have extracted the Toxic version of me from my body!"\n{'-' * 80}\n""")
    #read before continuing
    time.sleep(7)
    
    print (f"""
...As they start fighting you find a thunder gun and you point it at the two,
shooting at Toxic Rick will kill him, shooting in the middle will make them 
unite again in one person... I mean this is obvious, just basic particle physics.
    {'-' * 80}\n""")

    txr_a = ["A", "Kill", "Kill toxic rick", "First", "1", "One", "Shoot toxic rick"]
    txr_b = ["B", "Middle", "Shoot in the middle", "Second", "2", "Two", "In the middle"]

    target = ""

    #create loop for choice and repeating in case of diverse answer for TOXIC
    while True:
        target = input (prompt= f"""
What do you do?

    A) Kill Toxic Rick
    B) Shoot in the middle\n
        
""").capitalize()

        #if answer is wrong call retry option
        if target in txr_a:
            print (f"""
This did not work, Good Rick is dying too now, 
and you cannot really survive in space without him.\n{'-' * 80}""")
            retry()   
            break   
        #right choice break cycle TOXIC
        elif target in txr_b:
            print (f"""
Brave choice, and it worked! They are now the same person again, 
the old moody Rick is back.\n{'-' * 80}""")
            break

        #different answer repeats cycle TOXIC
        else:
            print (f"""
The gun does not work with these indications, choose one of the two options""")

    
    #final message for SECOND scenario RICK
    print (f"""
Merging with Toxic Rick was not acceptable for the Council of Ricks in the Citadel,
he committed too many galactic crimes before,they are now hunting you down, 
a bunch of them have actually found you already! Rick is starting the massacre, 
but three are about to approach you...
""")
    time.sleep(5)
    
    
    #final scene for SECOND scenario
    print (f"""
Rick throws a gun at you,

SHOOT THEM!\n""")
    #short loop to shoot 3 times with enter and WIN
    shot = 3

    while shot > 0:
        input (prompt= "Press <Enter> to shoot one Council Rick\n")
        shot -= 1

    print (f"""\n{'-' * 80}
Nice aim! You got them all...


                            YOU WON {name}!!!
                            
                   You have earned Rick's respect now.
{'-' * 80}""")
    exit(0)
    



#runs the multiverse scene (beginning)
multiv()


    
```
